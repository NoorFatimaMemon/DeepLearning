# **Self Organizing Map (SOM) or Kohonen Map or Self-organizing feature map (SOFM)**

SOM is an unsupervised learning technique used to map high-dimensional data onto a low-dimensional (typically two-dimensional) grid while preserving the topological properties of the input data. It's particularly useful for visualizing patterns in complex data.

# **Example Overview**

In this example, we'll create a synthetic dataset with two distinct clusters and apply the SOM to classify the data points. We'll also visualize the results using a scatter plot.

## **Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn_som.som import SOM # use to create and train the Self-Organizing Map.
```

## **Generating Synthetic Data**

```python
np.random.seed(0)
cluster_1 = np.random.normal(loc=[2, 2], scale=0.5, size=(100, 2))
cluster_2 = np.random.normal(loc=[7, 7], scale=0.5, size=(100, 2))
X = np.vstack((cluster_1, cluster_2))
```

cluster_1: Generates 100 data points centered around [2, 2] with a standard deviation of 0.5.

cluster_2: Generates 100 data points centered around [7, 7] with a standard deviation of 0.5.

np.vstack((cluster_1, cluster_2)): Vertically stacks the two clusters into a single array X, representing the dataset.

## **Defining our SOM Model**

```python
# Initializes a SOM with a 2x2 grid (2 rows and 2 columns) and 2-dimensional input data. 
som = SOM(m=2, n=2, dim=2, random_state=0)
```

# **Fitting the SOM Model**

```python
som.fit(X)
y_pred = som.predict(X)
```

# **Visualizing the Results**

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
plt.title('Clusters Identified by SOM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()
```
